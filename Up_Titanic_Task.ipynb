# 1. Prints a count of missing data for each column

```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv('/titanic.csv')
```

```python
titanic.head()
```

```python
titanic.isnull()
```

```python
titanic.isnull().any()
```

```python
titanic.isnull().sum()
```

```python
missing_age_values=titanic['Age'].isnull().sum()
missing_cabin_values=titanic['Cabin'].isnull().sum()
missing_embarked_values=titanic['Embarked'].isnull().sum()
print('Count of missing data in Age column is :', missing_age_values) 
print('Count of missing data in Cabin column is :', missing_cabin_values) 
print('Count of missing data in Embarked column is :', missing_embarked_values) 
```

# 2. Prints average age of all the people who survived



```python
avg_age=titanic[titanic['Survived'] == True]['Age'].mean()
print('Average age of people who survived is : ')
print(avg_age, 'years old')
```
